# Matplotlib

Matplotlib is a Python library used for creating data visualizations (graphs plots) in Python using code. It was created by John D. Hunter 

To use matplotlib, you need to install it and then import it.

Inside Matplotlib library is the Pyplot submodule which is what is used for plotting. 

Pyplot is imported as plt.

```python
import matplotlib
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

# Plotting a Line on Matplotlib

Draw a line from position (0,0) to position (6, 250)

```python
# Create two numpy arrays for the positions. 
# We will call the positions, xpoints and ypoints.

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

# now use the plot() function of pyplot to plot the axes
plt.plot(xpoints, ypoints)
plt.show()
```

# Plotting x and y points

The plot() function is used to draw points (markers) in a diagram. By default, the plot() function draws a line from point to point.

Plot() takes parameters for specifying points in the diagram.

Parameter 1 is an array containing the points on the x-axis

Parameter 2 is an array containing the points on the y- axis

If for example we need to plot a line from (1, 3) to (8, 10), we have to pass two arrays [1, 8] and [3, 10] to the plot function.

```python
# Example

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

# Plotting without 

To plot only the markers, you can use shortcut string notation parameter 'o' which means 'rings'

```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

# Plotting Multiple Points

You can plot as many points as you like but you need to make sure you have the same number of points on both axes.

```python
# Example
# Draw a line in a diagram from position (1, 3) to (2, 8) then (6, 1) and finally to position (8, 10)

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

# Defualt Values for The X-Axis

If no points are specified for the x-axis, they will get the default values 0,1,2,3,4, (etc. depending on the length of the y-points.

```python
# Example, when we take away xpoints from the above example.

ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

# Matplotlib Marker Types

Use the keyword argument marker to emphasize each point with a specified marker:


```python
# Example

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker= "o")
plt.show()
```

```python
# You can use different kinds of markers like star '*', diamond 'd', X 'x' and so much more
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker= "d")
plt.show()
```

# You can Use Format Strings to specify your marker as well as style your plot.

This involves using shortcut string notation parameter. This parameter is also called fmt, and is written with the syntax...

marker|line|color


```python
# Example, create a plot that has a star marker, dotted line, and green in color

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, '*:g')
plt.show()
```

# Marker Size

Use the keyword argument markersize or ms for short to set the size of the markers:

```python
# Example, Set the size of the markers to 20:

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker= 'o', ms= 20)
plt.show()
```

# Marker Color

Use the keyword argument markeredgecolor or mec to set the color of the edge of the markers:

# markerfacecolor or mfc

Use the keyword argument markerfacecolor or mfc to set the color inside the edge of the markers.

```python
# Let's set the edge color to red.

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker= "o", ms= 20, mec= 'r')
plt.show()
```

```python
# Set markerfacecolor

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker= "o", ms= 20, mfc= 'r')
plt.show()
```

```python
# Use both mec and mfc to set the marker to red by setting both to red


ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker= "o", ms= 20, mec= 'r', mfc= 'r')
plt.show()
```

#####  NB. Hexadecimal colors can be used as well.

# Lines in Matplotlib

Use the keyword argument linestyle or ls to change the style of plotted lines.

```python
# example:

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle='dotted')
plt.show()
```

```python
# dashed line


ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle= "dashed")
plt.show()
```

## Make The  Syntax Shorter.

You can make the syntax shorter.

1. linestyle can be written as ls

2. dotted can be written as :

3. dashed can be written as --

```python

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, ls= ":")
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, ls= "-.")
plt.show()
```

## Line Color

Use the keyword argument color or c for short to set the color of the line.

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, ls= ":", c= "r")
plt.show()
```

## Line Width

Use the keyword argument linewidth or the shorter lw to change the width of the line.

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, ls= "-", lw= "20")
plt.show()
```

# Matplotlib Labels and Title

You can create labels with pyplot using the xlabel() and ylabel() functions to set a label for the x- and y- axis.



```python
# Example.

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

# Create a Title for a Plot

Use the title() function to set a title for the plot in Pyplot

```python

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

# Set Font Properties for Title and Labels

Use fontdict parameter in xlabel(), ylabel(), and title() to set font properties for the title and labels

```python
# Example

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif', 'color':'blue', 'size':20}
font2 = {'family':'serif', 'color':'darkred', 'size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

### Position the Title

Use the loc parameter in the title() to position the title

loc can take the values 'left', 'right', and 'center'. Default value is 'center'

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data", loc= 'left')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.show()
```

Use the grid() function with Pyplot to add grid lines to the plot.

# Matplotlib Subplots

With the subplot() function you can draw multiple plots on the same canvas or plotting layout.

The subplot() function takes 3 arguments that describe the layout of the figure or drawing area.

The layout is organized in rows and columns. The first and second arguments of the subplot() function represent the row and column respectively.

The third argument represents the index of the current plot and specifies if it comes first, second, or third, etc.

```python
# Create two plots side by side 

#plot 1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

#### You can Draw as many plots as you want on one figure (layout), just describe the number of rows, columns, and the index of the plot.

# Matplotlib Scatter Plot

Use the scatter() function to draw a scatter plot with Pyplot.

The scatter() function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values of the y-axis.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x,y)
plt.show()
```

# ColorMap

A colormap is like a list of colors, where each color has a value that ranges from 0 to 100.

The colormap is called 'Viridis.

## How to Use the ColorMap

To use the colormap, you specify it with the keyword argument "cmap" and the value of the colormap, for example 'viridis' which is one of the built-in colormaps available in Matplotlib.

Then you create an array with values from 0 to 100, which represents one value for each of the point in the scatter plot.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='plasma')

# include the colormap in the drawing by including plt.colorbar() function
plt.colorbar()

plt.show()
```

## Size

You can change the size of the dots with the s argument.

Just like colors, make sure the array for sizes has the same length as the arrays for the x- axis- and y-axis:

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)

plt.show()
```

## Alpha

You can adjust the transparency of the dots with the alpha argument.

Just like colors, make sure the array for sizes has the same length as the arrays for the x- and y- axis.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
```

## Combine Color, Size, and Alpha

You can combine a colormap with different sizes on the dots. This is best visualized if the dots are transparent.

```python
# Create random arrays with 100 values for x-points, y-points, colors and sizes
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randint(100, size= (100))
y = np.random.randint(100, size= (100))
colors = np.random.randint(100, size= (100))
sizes =10*np.random.randint(100, size= (100))

plt.scatter(x, y, c= colors, s= sizes, alpha= 0.5, cmap= "nipy_spectral")

plt.colorbar()

plt.show()
```

# Matplotlib Pie Charts

With Pyplot, you can use the pie() function to draw pie charts:

Pie chart draws one pieces (called wedge) for each value in the array.

It plots the first wedge from the x- axis and move counterclockwise

### NB: 

The size of each wedge is determined by comparing the value with all the other values by using this formula.

The value divided by the sum of all values: x/sum(x)

```python
# Draw a pie chart.

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

## Labels

Add labels to the pie chart with the labels keyword parameter.

The labels parameter must be an array with one label for each wedge:

```python
y = np.array([35, 25, 25, 15])
piechart_labels = ['Apples', 'Bananas', 'Cherries', 'Dates']

plt.pie(y, labels= piechart_labels)
plt.show()
```

# Start Angle

Default start angle is at the x- axis, but you can change the start angle by specifying a startangle parameter.

The startangle parameter is defined with an angle in degrees, default angle is 0.

```python
# Start the first wedge at 90 degrees

y = np.array([35, 25, 25, 15])
piechart_labels = ['Apples', 'Bananas', 'Cherries', 'Dates']

plt.pie(y, labels= piechart_labels, startangle= 90)
plt.show()
```

# Explode

Use the explode keyword parameter to make one of the wedges stand out.

The explode parameter, when specified as not None, must be an array with one value for each wedge.

Each value represents how far from the center eacch wedge is displayed.

```python
# Pull the "Apples" 0.2 from the center of the pie.

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

# Shadow

Add a shadow to the pie chart by setting the shadows parameter to True:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

# Colors

Set the color of each wedge with the colors parameter.

The colors parameter, if specified, must be an array with one value for each wedge:

```python
y = np.array([35, 25, 25, 15])
mylabels = ['Apples', 'Bananas', 'Cherries', 'Dates']
mycolors = ['black', 'hotpink', 'b', '#4caf50']

plt.pie(y, labels= mylabels, colors= mycolors)
plt.show()
```

# Legend

Use the legend() function to add a list of explanation for each wedge.

```python
y = np.array([35, 25, 25, 15])
mylabels= ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels= mylabels)
plt.legend()
plt.show()
```

# Legend With Header

To add a header to the legend, add the title keyword parameter to the legend function.

```python
# Add a legend with a header

y = np.array([35, 25, 25, 15])
mylabels= ["Apples", "Bananas", "Cherries", "Dates"]
mycolors= ["#8DB600", "#ffe135", "#D2042D", "#4E3524"]

plt.pie(y, labels= mylabels, colors= mycolors)
plt.legend(title= "Four Sweet Fruits")
plt.show()
```
